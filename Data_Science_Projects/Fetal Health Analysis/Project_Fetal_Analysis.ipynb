##  List of Variables
```
FileName: of CTG examination
Date: of the examination
b: start instant
e: end instant
LBE: baseline value (medical expert)
LB: baseline value (SisPorto)
AC: accelerations (SisPorto)
FM: foetal movement (SisPorto)
UC: uterine contractions (SisPorto)
ASTV: percentage of time with abnormal short term variability (SisPorto)
mSTV: mean value of short term variability (SisPorto)
ALTV: percentage of time with abnormal long term variability (SisPorto)
mLTV: mean value of long term variability (SisPorto)
DL: light decelerations
DS: severe decelerations
DP: prolongued decelerations
DR: repetitive decelerations
Width: histogram width
Min: low freq. of the histogram
Max: high freq. of the histogram
Nmax: number of histogram peaks
Nzeros: number of histogram zeros
Mode: histogram mode
Mean: histogram mean
Median: histogram median
Variance: histogram variance
Tendency: histogram tendency: -1=left assymetric; 0=symmetric; 1=right assymetric
A: calm sleep
B: REM sleep
C: calm vigilance
D: active vigilance
SH: shift pattern (A or Susp with shifts)
AD: accelerative/decelerative pattern (stress situation)
DE: decelerative pattern (vagal stimulation)
LD: largely decelerative pattern
FS: flat-sinusoidal pattern (pathological state)
SUSP: suspect pattern
CLASS: Class code (1 to 10) for classes A to SUSP
NSP:- Normal=1; Suspect=2; Pathologic=3
```

````python
data1.info()
````

````python
data1.columns
````

````python
#splitting the data in input features and output features
y1 = data1['NSP']
y1 = y1.astype(np.int16)
X1 = data1.drop(['NSP'], axis = 1)
````

````python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to balance the classes
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X1, y1)

# Check the class distribution after resampling
print(y_resampled.value_counts())
````

Due to the Datasets imbalance, the Synthetic Minority Oversampling Technique (SMOTE) is employed to balance the Output class. SMOTE is used to prevent the ML Model from being overfit on Skewed classed.

````python
# Count the values of the "NSP" column
nsp_counts = y_resampled.value_counts()

# Plot the histogram
plt.figure(figsize=(4, 4))
nsp_counts.plot(kind='bar', color='skyblue')
plt.title('Count of NSP Classes after Balancing')
plt.xlabel('NSP Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
````

````python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_resampled, y_resampled, test_size = 0.2)
print(X_train1.shape, y_test1.shape)
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split

# Step 1: Create an instance of DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()

# Step 2: Fit the classifier to the training data
dt_classifier.fit(X_train1, y_train1)

# Step 3: Predict the labels for the training data
y_train_pred = dt_classifier.predict(X_train1)

print('################ For Training Data ##############')
print('\n')

# Step 4: Evaluate the performance of the classifier for training data
train_accuracy = accuracy_score(y_train1, y_train_pred)
print("Training Accuracy:", train_accuracy)

# Generate classification report for training data
train_report = classification_report(y_train1, y_train_pred)
print("Training Classification Report:")
print(train_report)

# Generate confusion matrix for training data
train_conf_matrix = confusion_matrix(y_train1, y_train_pred)
print("Training Confusion Matrix:")
print(train_conf_matrix)

print('\n')
print('################ For Testing Data ##############')
print('\n')
# Step 5: Predict the labels for the testing data
y_test_pred = dt_classifier.predict(X_test1)

# Step 6: Evaluate the performance of the classifier for testing data
test_accuracy_dt = accuracy_score(y_test1, y_test_pred)
print("Testing Accuracy:", test_accuracy_dt)

# Generate classification report for testing data
test_report = classification_report(y_test1, y_test_pred)
print("Testing Classification Report:")
print(test_report)

# Generate confusion matrix for testing data
test_conf_matrix = confusion_matrix(y_test1, y_test_pred)
print("Testing Confusion Matrix:")
print(test_conf_matrix)
````

````python
max_depths = range(1, 21)  # Testing depths from 1 to 20

train_accuracies = []
test_accuracies = []

for depth in max_depths:
    # Create Decision Tree Classifier with the specified depth
    dt_classifier = DecisionTreeClassifier(max_depth=depth)

    # Train the model
    dt_classifier.fit(X_train1, y_train1)

    # Predict on training set
    train_pred = dt_classifier.predict(X_train1)

    # Predict on testing set
    test_pred = dt_classifier.predict(X_test1)

    # Calculate training and testing accuracies
    train_accuracy = accuracy_score(y_train1, train_pred)
    test_accuracy = accuracy_score(y_test1, test_pred)

    # Append accuracies to lists
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)

# Plot the training and testing accuracies
plt.figure(figsize=(10, 6))
plt.plot(max_depths, train_accuracies, marker='o', label='Training Accuracy')
plt.plot(max_depths, test_accuracies, marker='o', label='Test Accuracy')
plt.title('Decision Tree Performance vs. Max Depth')
plt.xlabel('Max Depth')
plt.ylabel('Accuracy')
plt.xticks(max_depths)
plt.legend()
plt.grid(True)
plt.show()
````

````python
from sklearn.ensemble import RandomForestClassifier

# Step 1: Create an instance of RandomForestClassifier
rf_classifier = RandomForestClassifier()

# Step 2: Fit the classifier to the training data
rf_classifier.fit(X_train1, y_train1)

# Step 3: Predict the labels for the training data
y_train_pred = rf_classifier.predict(X_train1)

print('################ For Training Data ##############')
print('\n')

# Step 4: Evaluate the performance of the classifier for training data
train_accuracy = accuracy_score(y_train1, y_train_pred)
print("Training Accuracy:", train_accuracy)

# Generate classification report for training data
train_report = classification_report(y_train1, y_train_pred)
print("Training Classification Report:")
print(train_report)

# Generate confusion matrix for training data
train_conf_matrix = confusion_matrix(y_train1, y_train_pred)
print("Training Confusion Matrix:")
print(train_conf_matrix)

print('\n')
print('################ For Testing Data ##############')
print('\n')

# Step 5: Predict the labels for the testing data
y_test_pred = rf_classifier.predict(X_test1)

# Step 6: Evaluate the performance of the classifier for testing data
test_accuracy_rf = accuracy_score(y_test1, y_test_pred)
print("Testing Accuracy:", test_accuracy_rf)

# Generate classification report for testing data
test_report = classification_report(y_test1, y_test_pred)
print("Testing Classification Report:")
print(test_report)

# Generate confusion matrix for testing data
test_conf_matrix = confusion_matrix(y_test1, y_test_pred)
print("Testing Confusion Matrix:")
print(test_conf_matrix)
````

````python
# Extract and display important features
importance = rf_classifier.feature_importances_
features = X_train1.columns
feature_importance = dict(zip(features, importance))
sorted_feature_importance = sorted(feature_importance.items(), key=lambda x: x[1], reverse=True)

# Extracting feature names and importances
feature_names = [feature[0] for feature in sorted_feature_importance]
importances = [feature[1] for feature in sorted_feature_importance]

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.barh(feature_names, importances, color='skyblue')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.title('Feature Importances')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest feature on top
plt.show()
````

````python
# Define a range of values for the number of trees
n_estimators_range = range(1, 101, 5)  # From 1 to 100 with step size 5

train_accuracies = []
test_accuracies = []

for n_estimators in n_estimators_range:
    # Create Random Forest Classifier with the specified number of trees
    clf_RF = RandomForestClassifier(n_estimators=n_estimators, random_state=42)

    # Train the model
    clf_RF.fit(X_train1, y_train1)

    # Predict on training set
    train_pred = clf_RF.predict(X_train1)

    # Predict on testing set
    test_pred = clf_RF.predict(X_test1)

    # Calculate training and testing accuracies
    train_accuracy = accuracy_score(y_train1, train_pred)
    test_accuracy = accuracy_score(y_test1, test_pred)

    # Append accuracies to lists
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)

# Plot the training and testing accuracies
plt.figure(figsize=(10, 6))
plt.plot(n_estimators_range, train_accuracies, marker='o', label='Training Accuracy')
plt.plot(n_estimators_range, test_accuracies, marker='o', label='Validation Accuracy')
plt.title('Random Forest Performance vs. Number of Trees')
plt.xlabel('Number of Trees')
plt.ylabel('Accuracy')
plt.xticks(n_estimators_range)
plt.legend()
plt.grid(True)
plt.show()
````

````python
from sklearn.svm import SVC

# Step 1: Create an instance of SVM classifier
svm_classifier = SVC(kernel='linear', decision_function_shape='ovr')  # Using linear kernel and OvR strategy

# Step 2: Fit the classifier to the training data
svm_classifier.fit(X_train1, y_train1)

# Step 3: Predict the labels for the training data
y_train_pred = svm_classifier.predict(X_train1)

# Step 4: Evaluate the performance of the classifier for training data
train_accuracy = accuracy_score(y_train1, y_train_pred)
print("Training Accuracy:", train_accuracy)

# Generate classification report for training data
train_report = classification_report(y_train1, y_train_pred)
print("Training Classification Report:")
print(train_report)

# Generate confusion matrix for training data
train_conf_matrix = confusion_matrix(y_train1, y_train_pred)
print("Training Confusion Matrix:")
print(train_conf_matrix)

# Step 5: Predict the labels for the testing data
y_test_pred = svm_classifier.predict(X_test1)

# Step 6: Evaluate the performance of the classifier for testing data
test_accuracy_svm = accuracy_score(y_test1, y_test_pred)
print("Testing Accuracy:", test_accuracy_svm)

# Generate classification report for testing data
test_report = classification_report(y_test1, y_test_pred)
print("Testing Classification Report:")
print(test_report)

# Generate confusion matrix for testing data
test_conf_matrix = confusion_matrix(y_test1, y_test_pred)
print("Testing Confusion Matrix:")
print(test_conf_matrix)
````

````python
# Define a range of values for the regularization parameter (C)
C_values = np.logspace(-3, 3, 7)  # From 0.001 to 1000

train_accuracies = []
test_accuracies = []

for C in C_values:
    # Create SVM classifier with the specified regularization parameter
    svm_clf = SVC(C=C, kernel='linear', decision_function_shape='ovr')

    # Fit the classifier to the training data
    svm_clf.fit(X_train1, y_train1)

    # Predict on training set
    y_train_pred = svm_clf.predict(X_train1)

    # Predict on testing set
    y_test_pred = svm_clf.predict(X_test1)

    # Calculate training and testing accuracies
    train_accuracy = accuracy_score(y_train1, y_train_pred)
    test_accuracy = accuracy_score(y_test1, y_test_pred)

    # Append accuracies to lists
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)

# Plot the training and testing accuracies
plt.figure(figsize=(10, 6))
plt.plot(C_values, train_accuracies, marker='o', label='Training Accuracy')
plt.plot(C_values, test_accuracies, marker='o', label='Validation Accuracy')
plt.title('SVM Performance vs. Regularization Parameter (C)')
plt.xlabel('Regularization Parameter (C)')
plt.ylabel('Accuracy')
plt.xscale('log')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# Preprocess data for SVM (scaling)
def preprocess_for_svm(new_instance):
    scaler = StandardScaler()
    new_instance_scaled = scaler.fit_transform(new_instance)
    return new_instance_scaled

# Preprocess data for ANN (encoding categorical variables, scaling)
# Assuming categorical variables are one-hot encoded and scaling is already applied
def preprocess_for_ann(new_instance):
    return new_instance
    # scaler = StandardScaler()
    # new_instance_scaled = scaler.fit_transform(new_instance)
    # return new_instance_scaled
````

````python
# Predict result for Decision Tree
def predict_with_decision_tree(model, new_instance, labels):
    result = model.predict(new_instance)
    return [labels[i-1] for i in result]

# Predict result for Random Forest
def predict_with_random_forest(model, new_instance, labels):
    result = model.predict(new_instance)
    return [labels[i-1] for i in result]

# Predict result for SVM
def predict_with_svm(model, new_instance, labels):
    preprocessed_instance_svm = preprocess_for_svm(new_instance)
    result = model.predict(preprocessed_instance_svm)
    return [labels[i-1] for i in result]

# Predict result for ANN
def predict_with_ann(model, new_instance, labels):
    preprocessed_instance_ann = preprocess_for_ann(new_instance)
    probabilities = model.predict(preprocessed_instance_ann)
    predicted_label_index = np.argmax(probabilities)
    predicted_label = labels[predicted_label_index]
    return [predicted_label]
````
